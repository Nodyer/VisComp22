Manipulação de imagens com OpenCV
================

#### Leitura e exibição em janela de uma imagem

Importando a biblioteca:

```python
import cv2
```

Realizando a leitura da imagem:

```python
img = cv2.imread('onca.jpg') # leitura da imagem
cv2.imshow("Onça", img) # mostra a image

if cv2.waitKey(0) == 27: # esc para fechar a janela
    cv2.destroyAllWindows()
```

#### Explorando o segundo parâmetro da função `cv2.imread()`

A função `cv2.imread()` que especifica o formato da imagem que deseja ler.
- cv2.IMREAD_COLOR (ou 1): Carrega as cores da imagem.
- cv2.IMREAD_GRAYSCALE (ou 0): Carrega a imagem com escalas de cinza. 
- cv2.IMREAD_UNCHANGED (ou -1): Carrega a imagem incluindo o canal alfa.

```python
img = cv2.imread('onca.jpg',0)
cv2.imshow("Onça", img)

if cv2.waitKey(0) == 27:
    cv2.destroyAllWindows()
```

```python
img = cv2.imread('onca.jpg',1)
cv2.imshow("Onça", img)

if cv2.waitKey(0) == 27:
    cv2.destroyAllWindows()
```

```python
img = cv2.imread('onca.jpg',-1)
cv2.imshow("Onça", img)

if cv2.waitKey(0) == 27:
    cv2.destroyAllWindows()
```

#### Rotação de imagem

90 graus:

```python
img = cv2.imread('onca.jpg')

img_rot = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE) # rotaciona a imagem em 90 graus
  
cv2.imshow("90 graus", img_rot)

if cv2.waitKey(0) == 27:
    cv2.destroyAllWindows()
```

180 graus:

```python
img = cv2.imread('onca.jpg')

img_rot = cv2.rotate(img, cv2.ROTATE_180) # rotaciona a imagem em 180 graus
  
cv2.imshow("180 graus", img_rot)

if cv2.waitKey(0) == 27:
    cv2.destroyAllWindows()
```
