Transformada de Hough Probabilístico
===

Importando as bibliotecas:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carregando a image
img = cv2.imread(cv2.samples.findFile('fotolia.jpg'))
img_lines = np.copy(img)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,150,apertureSize = 3)

# Detectando as linhas
lines = cv2.HoughLinesP(edges,1,np.pi/180,90,minLineLength=10,maxLineGap=50)

# Desenhando as linhas detectadas
for line in lines:
    x1,y1,x2,y2 = line[0]
    cv2.line(img_lines,(x1,y1),(x2,y2),(0,0,255),2)

# Exibição dos resultados
plt.figure(figsize=(20,7))
plt.subplot(131), plt.title("Original"), plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)), plt.axis('off')
plt.subplot(132), plt.title("Canny Edge Detection"), plt.imshow(edges, 'gray'), plt.axis('off')
plt.subplot(133), plt.title("Probabilistic Hough Line Transform"), plt.imshow(cv2.cvtColor(img_lines,cv2.COLOR_BGR2RGB)), plt.axis('off')
plt.show()
```
