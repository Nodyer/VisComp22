```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregando as imagens
imgs = [
  cv2.imread('fog-oceano/fog-001-800x600.jpg'),
  cv2.imread('fog-oceano/fog-002-800x600.jpg'),
  cv2.imread('fog-oceano/fog-04.jpg'),
  cv2.imread('fog-oceano/fog9.jpg')
]

# Realizando uma cpia
original = imgs.copy()

# Para RGB
xp = [0, 64, 128, 192, 255]
fp = [0, 16, 128, 240, 255]
x = np.arange(256)
table = np.interp(x, xp, fp).astype('uint8')

for i in range(len(imgs)):
    imgs[i] = cv2.LUT(imgs[i], table)

# Para HSV
imgs_hsv = []

for i in imgs:
  rgb2hsv = cv2.cvtColor(i, cv2.COLOR_BGR2HSV)
  imgs_hsv.append(i)

for i in range(len(imgs_hsv)):
  imgs_hsv[i] = cv2.LUT(imgs_hsv[i], table)

# Exibio dos resultados

i = 0
j = 0

plt.figure(figsize=(7,10))
while i < 4:
  plt.subplot(4,3, 1 + j), plt.imshow(cv2.cvtColor(original[i], cv2.COLOR_BGR2RGB)), plt.title("Original"), plt.axis("off")
  plt.subplot(4,3, 1 + j + 1), plt.imshow(cv2.cvtColor(imgs[i], cv2.COLOR_BGR2RGB)), plt.title("RGB"), plt.axis("off")
  plt.subplot(4,3, 1 + j + 2), plt.imshow(cv2.cvtColor(imgs_hsv[i], cv2.COLOR_HSV2RGB)), plt.title("HSV"), plt.axis("off")
  i+=1
  j+=3
  
plt.show()
```
